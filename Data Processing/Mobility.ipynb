```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mobilityPath = "Resources/Mobility_Report_Feb-May2020.csv"


# Read the mouse data and the study results
mobility = pd.read_csv(mobilityPath)

# Display the data table for preview
mobility.head()
```

```python

retail = mobility.groupby("county").sum()["retail_and_recreation_percent_change_from_baseline"].astype(int)
grocery_pharmacy = mobility.groupby("county").sum()["grocery_and_pharmacy_percent_change_from_baseline"].astype(int)
parks = mobility.groupby("county").sum()["parks_percent_change_from_baseline"].astype(int)
transit_stations = mobility.groupby("county").sum()["transit_stations_percent_change_from_baseline"].astype(int)
workplace = mobility.groupby("county").sum()["workplaces_percent_change_from_baseline"].astype(int)
residential = mobility.groupby("county").sum()["residential_percent_change_from_baseline"].astype(int)

summary_table = pd.DataFrame({"retail_and_recreation_percent_change_from_baseline":retail,
                              "grocery_and_pharmacy_percent_change_from_baseline":grocery_pharmacy,
                              "parks_percent_change_from_baselines":parks,
                              "transit_stations_percent_change_from_baseline":transit_stations,
                             "workplaces_percent_change_from_baseline":workplace,
                             "residential_percent_change_from_baseline":residential})

summary_table
```

```python
summary_table.to_csv('Resources/mobility.csv')
```
