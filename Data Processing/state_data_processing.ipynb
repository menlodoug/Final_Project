```python
import pandas as pd
from pathlib import Path
```

```python
from sqlalchemy import create_engine
from config import db_password
```

```python
# file path
db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/covid_19_ca"
engine = create_engine(db_string)
```

```python
def parser(x):
	return pd.to_datetime(x)

case_death_data = pd.read_csv(Path('./Resources/us_states_daily.csv'), header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
case_death_data.head()
```

```python
# Get California data only
ca_case_df = case_death_data.loc[case_death_data['state'] == 'California']
ca_case_df = ca_case_df.drop(columns=["state", "fips", "cases", "deaths"])
ca_case_df.sample(10)
```

```python
# save the cases data to the postgress
ca_case_df.to_sql(name='california_state_cases', con=engine, if_exists='replace')
```

```python
# Global_Mobility_Report.csv
mobility_raw = pd.read_excel('./Resources/Global_Mobility_Report.xlsx', index_col=0) 
```

```python
# transform the date to datetime
import datetime as dt
mobility_raw['datetime'] = pd.to_datetime(mobility_raw['date'])
```

```python
# set the date as index
mobility_raw = mobility_raw.set_index(['datetime'])
```

```python
# remove the country, date columns
ca_mobility_raw = mobility_raw.drop(columns=["date"])
```

```python
# save the mobility data to the postgress
ca_mobility_raw.to_sql(name='california_state_mobility', con=engine, if_exists='replace')
```
