```python
# Get California data only
ca_df = data.loc[data['state'] == 'California']
ca_df.head()
```

```python
# Showing the autocorrelation for cases
cases = ca_df.drop(["state", "fips", "deaths"], axis=1)
```

## Make the data stationary

```python
# using 32 as window for rolling mean since up to that point the numbers have a slow increase rate
```

```python
def get_stationarity(timeseries, key):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=32).mean()
    rolling_std = timeseries.rolling(window=32).std()
    
    # rolling statistics plot
    original = pyplot.plot(timeseries, color='blue', label='Original')
    mean = pyplot.plot(rolling_mean, color='red', label='Rolling Mean')
    std = pyplot.plot(rolling_std, color='black', label='Rolling Std')
    pyplot.legend(loc='best')
    pyplot.title('Rolling Mean & Standard Deviation')
    pyplot.show(block=False)
    
    # Dickeyâ€“Fuller test:
    result = adfuller(timeseries[key])
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```

```python
# check if the data is stationary using the rolling statistics
get_stationarity(cases, "cases")
```

The rolling mean and rolling standard deviation increase with time. Therefore, we can conclude that this time series is not stationary.

The ADF Statistic is far from the critical values and the p-value is greater than the threshold (0.05).
Thus, we can conclude that the time series is not stationary.

```python
# use the log to lower the rate at which rolling mean increases.
cases_log = np.log(cases)
pyplot.plot(cases_log)
```

### Appling transformation to the time series to render it stationary

```python
# Substracting the rolling mean
cases_rolling_mean = cases_log.rolling(window=30).mean()
cases_log_minus_mean = cases_log - cases_rolling_mean
cases_log_minus_mean.dropna(inplace=True)
get_stationarity(cases_log_minus_mean, "cases")
```

```python
# Appling exponential decay 
rolling_mean_exp_decay = cases_log.ewm(halflife=1, min_periods=0, adjust=True).mean()
cases_log_exp_decay = cases_log - rolling_mean_exp_decay
get_stationarity(cases_log_exp_decay, "cases")
```

```python
# applying time shifting
cases_log_shift = cases_log - cases_log.shift()
cases_log_shift.dropna(inplace=True)
get_stationarity(cases_log_shift, "cases")
```

```python
# Auto Correlation Function (ACF) to determine the optimal number of MA terms
model = ARIMA(cases_log_shift, order=(2,1,2))
results = model.fit(disp=-1)
pyplot.plot(cases_log_shift)
pyplot.plot(results.fittedvalues, color='red')
```

```python
results.summary()
```

```python
predictions_ARIMA_diff = pd.Series(results.fittedvalues, copy=True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_log = pd.Series(cases_log['cases'].iloc[0], index=cases_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value=0)
predictions_ARIMA = np.exp(predictions_ARIMA_log)
pyplot.plot(cases)
# pyplot.plot(predictions_ARIMA)
```
