## Make the data stationary

```python
# using 32 as window for rolling mean since up to that point the numbers have a slow increase rate
```

```python
def get_stationarity(timeseries, key):
    
    # rolling statistics
    rolling_mean = timeseries.rolling(window=10).mean()
    rolling_std = timeseries.rolling(window=10).std()
    
    # rolling statistics plot
    original = pyplot.plot(timeseries, color='blue', label='Original')
    mean = pyplot.plot(rolling_mean, color='red', label='Rolling Mean')
    std = pyplot.plot(rolling_std, color='black', label='Rolling Std')
    pyplot.legend(loc='best')
    pyplot.title('Rolling Mean & Standard Deviation')
    pyplot.show(block=False)
    
    # Dickey–Fuller test:
    result = adfuller(timeseries[key])
    print('ADF Statistic: {}'.format(result[0]))
    print('p-value: {}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {}'.format(key, value))
```

```python
# check if the data is stationary using the rolling statistics
get_stationarity(cases, "case_day")
```

The rolling mean and rolling standard deviation increase with time. Therefore, we can conclude that this time series is not stationary.

The ADF Statistic is far from the critical values and the p-value is greater than the threshold (0.05).
Thus, we can conclude that the time series is not stationary.

```python
# use the log to lower the rate at which rolling mean increases.
cases_log = np.log(cases)
pyplot.plot(cases_log)
```

```python
cases_log = cases_log.replace([np.inf, -np.inf], 0)
# maybe not using this 
```

### Appling transformation to the time series to render it stationary

```python
# Substracting the rolling mean
cases_rolling_mean = cases_log.rolling(window=40).mean()
cases_minus_mean = cases_log - cases_rolling_mean
cases_minus_mean.dropna(inplace=True)
get_stationarity(cases_minus_mean, "case_day")
```

```python
# Appling exponential decay 
rolling_mean_exp_decay = cases_log.ewm(halflife=1, min_periods=1, adjust=True).mean()
cases_exp_decay = cases_log - rolling_mean_exp_decay
cases_exp_decay.dropna(inplace=True)
get_stationarity(cases_exp_decay, "case_day")
```

```python
# applying time shifting
cases_shift = cases_log - cases_log.shift()
cases_shift.dropna(inplace=True)
get_stationarity(cases_shift, "case_day")
```
