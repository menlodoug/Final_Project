```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:.2f}'.format
```

```python
apple_df = pd.read_csv('applemobilitytrends-2020-06-07.csv')
apple_df
```

```python
apple = pd.melt(apple_df, 
            id_vars='County', 
            value_vars=list(apple_df.columns[1:155]), # list of days of the week
            var_name='Date', 
            value_name='Index')
apple
```

```python
apple.to_csv('apple.csv')
```

```python
states_df = pd.read_csv('us-states.csv')
states_df
```

```python
states_df['case_day'] = states_df.groupby('state')['cases'].diff().fillna(0)
states_df
```

```python
states_df['death_day'] = states_df.groupby('state')['deaths'].diff().fillna(0)
states_df.to_csv('us-states.csv')
```
